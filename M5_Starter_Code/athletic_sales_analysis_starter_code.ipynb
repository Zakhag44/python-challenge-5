```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python

from pathlib import Path

# Read the CSV files into DataFrames.
path_2020 = Path('Resources/athletic_sales_2020.csv')
path_2021 = Path('Resources/athletic_sales_2021.csv')
```

```python

import pandas as pd
# Display the 2020 sales DataFrame
path_2020_df = pd.read_csv(path_2020)
path_2020_df.head()
```

```python
# Display the 2021 sales DataFrame
path_2021_df = pd.read_csv(path_2021)
path_2021_df.head()
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
path_2020_df.dtypes
```

```python
# Check the 2021 sales data types.
path_2021_df.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_df = pd.concat([path_2020_df, path_2021_df], axis="rows").reset_index(drop=True)
combined_df.head()
```

```python
# Check if any values are null.
combined_df.isnull().sum()
```

```python
# Check the data type of each column
combined_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_df["invoice_date"] = pd.to_datetime(combined_df["invoice_date"])
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
products_sold = combined_df.groupby(["region", "state", "city"]).agg({"product": "count"}).reset_index()
# Rename the sum to "Total_Products_Sold".
products_sold.rename(columns={"product": "Total_Products_Sold"}, inplace=True)

# Show the top 5 results.
products_sold.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold = combined_df.groupby(["region", "state", "city"]).agg({"product": "count"}).reset_index()

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold.rename(columns={"product": "Total_Products_Sold"}, inplace=True)


# Show the top 5 results.
products_sold.head()
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales = combined_df.groupby(["region", "state", "city"]).agg({"total_sales": "sum"}).reset_index()
# Rename the "total_sales" column to "Total Sales"
total_sales.rename(columns={"total_sales": "Total_Sales"}, inplace=True)

# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales = combined_df.groupby(["region", "state", "city"]).agg({"total_sales": "sum"}).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
total_sales.head()
```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
total_sales = combined_df.groupby(["retailer", "region", "state", "city"]).agg({"total_sales": "sum"}).reset_index()
total_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
total_sales.head()
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
total_sales = combined_df.groupby(["retailer", "region", "state", "city"]).agg({"total_sales": "sum"}).reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
total_sales.rename(columns={"total_sales": "Total Sales"}, inplace=True)

# Show the top 5 results.
total_sales.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear = combined_df[combined_df['product'] == "Women's Athletic Footwear"]
womens_footwear.head()
                                          
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales = womens_footwear.groupby(["retailer", "region", "state", "city"]).agg({"units_sold": "sum"}).reset_index()
womens_footwear_sales.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)

# Show the top 5 results.
womens_footwear_sales.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_sales = womens_footwear.groupby(["retailer", "region", "state", "city"]).agg({"units_sold": "sum"}).reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"}, inplace=True)
# Show the top 5 results.
womens_footwear_sales.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column as the index, and the "total_sales" as the values.
pivot_table = combined_df.pivot_table(index='invoice_date', values='total_sales')

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
pivot_table
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
resampled_pivot_table = pivot_table.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
resampled_pivot_table.sort_values(by='Total Sales', ascending=True, inplace=True)

# Show the table.
resampled_pivot_table
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
resampled_pivot_table = pivot_table.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
resampled_pivot_table.sort_values(by='Total Sales', ascending=True, inplace=True)

# Show the table.
resampled_pivot_table
```
